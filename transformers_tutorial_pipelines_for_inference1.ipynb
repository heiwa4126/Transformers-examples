```python
from transformers import pipeline
```

## 自動音声認識 (Automated Speech Recognition : ASR)

```python
transcriber = pipeline(task="automatic-speech-recognition")
```

```python
transcriber("https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac")
```

あまりいい結果とはいえない、もっとよいモデルに変えてみよう。

```python
transcriber2 = pipeline(model="openai/whisper-large-v2")
```

```python
transcriber2("https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac")
```

```python
transcriber(
    [
        "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac",
        "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/1.flac",
    ]
)
```

```python
transcriber2(
    [
        "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/mlk.flac",
        "https://huggingface.co/datasets/Narsil/asr_dummy/resolve/main/1.flac",
    ]
)
```
